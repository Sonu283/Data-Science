```python
data = {
    "Name" : ["Alice", "Bob", "Charlie", "David", "Eve", "Alice"],
    "Age" : [25, 30, 35, np.nan, 29, 25],
    "Department" : ["HR", "IT", "Finance", "IT", "HR", "HR"],
    "Salary" : [50000, 60000, 70000, 62000, np.nan, 50000]
}
```

```python
# Broadcasting
df['Salary'] = df['Salary'] + 5000
```

```python
df['Salary']
```

```python
# Renaming Columns:
df.rename(columns = {"Department" : "Dept"}, inplace = True)
```

```python
df
```

```python
df["Dept"].unique()
```

```python
df["Dept"].value_counts()
```

```python
df["Promoted Salary"] = df["Salary"] * 10
```

```python
df
```

```python
#Data Cleaning
df.isnull().sum()
```

```python
df.dropna(how = 'any') # any row that had any null value
```

```python
df.dropna(how = 'all') #if all the values in any row are null then we  drop
```

```python
df["Age"].fillna(df["Age"].mean()) # also we can use inplace = True
```

```python
df["Salary"].fillna(df["Salary"].median())
```

```python
df["Age"].fillna(method = "ffill") #forward fill
```

```python
df["Age"].fillna(method = "bfill")# backward fill
```

```python
df["Name"] = df["Name"].replace("Charlie", "Rose")
```

```python
df
```

```python
# Duplicates:

df_dup = df[df.duplicated(keep = "first")]
df_dup
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
#invalid Values
# Lambda -> python

df["Promoted Salary"] = df["Promoted Salary"].apply(lambda x: x/10 if x > 650000 else x)
```

```python
df
```

```python
name = "alice_fernandes"
df[["first_name","last_name"]] = df["name"].str.split("_")  #it will not run because of my data but you can run by creating your own data (demo)
```

```python
# apply and lembda
def multiplying_age(x):
    return x*2

df["Age"] = df["Age"].apply(multiplying_age)
```

```python
df
```

```python
df["Age"] = df["Age"].apply(lambda x: x/2)
```

```python
df
```

```python
#Joins and Merges

department_info = {
    "Dept" : ["HR", "IT", "Finance"],
    "Location" : ["New York", "San Francisco", "Chicago"],
    "Manager" : ["Laura", "Steve", "Naina"]
}

df2 = pd.DataFrame(department_info)
df2
```

```python
#Concat
pd.concat([df, df2])
```

```python
pd.concat([df, df2], axis=1)
```

```python
pd.merge(df,df2, on = "Dept")
```
