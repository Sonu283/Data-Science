```python
df.head()
```

```python
df.info()
```

```python
# Pie Chart
count = df['Dept'].value_counts()
count
```

```python
plt.pie(count, labels=count.index, autopct='%1.2f%%', explode=(0, 0.1, 0.2))
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

```python
# Count Plot
plt.bar(count.index, count, color=['red', 'blue', 'green'])
plt.show()
```

```python
# Bivariate - Numerical vs Categorical
```

```python
hr_salary = df[df['Dept']=='HR']['Salary']
hr_salary
```

```python
it_salary = df[df['Dept'] == 'IT']['Salary']
it_salary
```

```python
fin_salary = df[df['Dept'] == 'Finance']['Salary']
fin_salary
```

```python
# Box Plot
plt.boxplot([hr_salary, it_salary, fin_salary], labels=['HR', 'IT', 'Finance'])
plt.show()
```

```python
# Pie Chart

salary_by_dept = df.groupby('Dept')['Salary'].sum()
salary_by_dept
```

```python
plt.pie(salary_by_dept, labels=salary_by_dept.index, autopct='%1.2f', shadow=True, explode=(0.05,0.05,0.05))
plt.axis('equal')
plt.show()
```

```python
# Bar Plot

hr_mean = sum(hr_salary)/len(hr_salary)
it_mean = sum(it_salary)/len(it_salary)
fin_mean = sum(fin_salary)/len(fin_salary)
```

```python
plt.bar(['HR','IT','Finance'],[hr_mean, it_mean, fin_mean], color=['red', 'blue', 'green'])
plt.grid()
plt.show()
```

```python
# Multivariate Analysis: 3 Numerical Variables
```

```python
df.head()
```

```python
df['Experience'] = [2, 5, 3, 7, 4, 10, 8, 2, 3, 6, 9, 5, 4, 7, 4, 11, 12, 2, 1, 10 ]
df.head()
```

```python
# Bubble Plot

plt.scatter(df['Age'],df['Salary'], s=df['Experience']*30, color='blue', edgecolor='black')
plt.title('Salary vs Age with Experience as Bubble Size')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

```python
# 2 Numerical and 1 Categorical column

plt.scatter(df['Age'],df['Salary'], c=df['Dept'].map({'HR':'red', 'IT':'blue', 'Finance':'green'}))
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary vs Age colored by Dept')
plt.show()
```

```python
color = {'HR':'red', 'IT':'blue', 'Finance':'green'}

for dept,color in color.items():
    df_dept = df[df['Dept'] == dept]
    plt.scatter(df_dept['Age'], df_dept['Salary'], c=color, label=dept)

plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Salary vs Age colored by Dept')
plt.legend()
plt.show()
```

```python
# Object Oriented API
```

```python
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

# Line Plot
axs[0].plot(sort_age['Age'], df['Salary'], marker ='*', linewidth=2, markersize=2, color='red')
axs[0].grid()
axs[0].set_title('Line Plot')
axs[0].set_xlabel('Age')    
axs[0].set_ylabel('Salary')

# histogram
axs[1].hist(df['Salary'], bins=5, color='blue')
axs[1].set_title('Histogram')
axs[1].set_xlabel('Salary')
axs[1].set_ylabel('Frequency')

# Box Plot
axs[2].boxplot(df['Salary'])
axs[2].set_title('Box Plot')
axs[2].set_xlabel('Salary')

plt.savefig('MultiplePlots.png')
plt.show()
```

```python
data2 = {
    'Years': [2015, 2016, 2017, 2018],
    'Sales': [200, 250, 300, 350],
    'Profit': [50, 70, 90, 120],
    "Expenses": [100, 120, 150, 180]
}
```

```python
df2 = pd.DataFrame(data2)
```

```python
plt.plot(df2['Years'], df2['Sales'], marker='o', label='Sales')
plt.plot(df2['Years'], df2['Profit'], marker='s', label='Profit')
plt.plot(df2['Years'], df2['Expenses'], marker='^', label='Expenses')

plt.title('Financial Analysis Over Years')
plt.xlabel('Years')
plt.ylabel('Amount')
plt.legend()
plt.show()
```
